**목표: X-ray 이미지를 불러와 기본 통계(해상도, 픽셀 분포)와 원본 이미지를 시각화.**

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io
```

```python
image_path = '/content/drive/MyDrive/Medical_CV/X-ray/X-ray 이미지/normal_10.jpg'  # 이미지 경로 설정
xray_image = io.imread(image_path, as_gray=True)  # 이미지를 흑백으로 로드
```

```python
# 이미지 크기와 채널 수 확인
print("Image Shape:", xray_image.shape)
```

```python

# 픽셀 값 분포 분석 (히스토그램)
plt.figure(figsize=(10, 5))
plt.hist(xray_image.ravel(), bins=256, range=(0.0, 1.0), fc='black', ec='black')
plt.title('Pixel Intensity Distribution')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# 원본 이미지 시각화
plt.figure(figsize=(7, 7))
plt.imshow(xray_image, cmap='gray')
plt.title("Original X-ray Image")
plt.axis('off')
plt.show()
```

```python
# 요약
print("Image Summary:")
print(f"- Resolution: {xray_image.shape[0]} x {xray_image.shape[1]}")
print(f"- Pixel Intensity Range: {xray_image.min()} to {xray_image.max()}")
```
